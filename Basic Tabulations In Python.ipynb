## loc and iloc

**loc** selects rows and columns with specific ***labels***.

but our data does not have labels, so the label will be their default index positions

```python
# select rows with certain labels ( note the index will be the labels as the data is not labelled)
Cancer_Data.loc[[1,2,35]] # select rows with index 1, 2 and 35
```

```python
# select rows and columns with certain labels
Cancer_Data.loc[[1,2,35,], ["age", "gender", "bmi"]] # note the label for a column is the name of the column
```

```python
# we can select a range using :
Cancer_Data.loc[5:10, :"bmi"] # rows index no. 1 to 10 of the first column to column "bmi"
```

```python
Cancer_Data.loc[5, "bmi"] # to select the fifth row of the column
```

**iloc** selects rows and columns at specific ***integer position***

```python
# select rows 4 and 5 bases on their index (not labels, iloc doesn't use labels)
Cancer_Data.iloc[4:6] # note it will exclude 6
```

```python
Cancer_Data.iloc[4:6, :4] # select rows 4 to 5(will exclude 6) and the first column to the 3rd(will exclude 4)
```

```python
Cancer_Data.iloc[:5, [1,2,3]] # first five rows in column, 1,2,3
```

###### note if we are selcecting multiple rows or columns without using range, we include square bracket
